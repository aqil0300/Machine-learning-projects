```python
q1=df['Prior_purchases'].quantile(0.25)
q3=df['Prior_purchases'].quantile(0.75)
iqr=q3-q1
lower=q1-(iqr*1.5)
upper=q3+(iqr*1.5)

for i in df['Prior_purchases']:
  if i<lower or i>upper:
    df['Prior_purchases']=df['Prior_purchases'].replace(i,np.mean(df['Prior_purchases']))
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
df['Product_importance']=label.fit_transform(df['Product_importance'])
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
model=KNeighborsClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
knn=accuracy_score(y_test,y_pred)
classification_report(y_test,y_pred)
```

```python
model2=SVC()
model2.fit(x_train,y_train)
y_pred=model2.predict(x_test)
svc=accuracy_score(y_test,y_pred)
classification_report(y_test,y_pred)
```

```python
model3=DecisionTreeClassifier()
model3.fit(x_train,y_train)
y_pred=model3.predict(x_test)
dtc=accuracy_score(y_test,y_pred)
classification_report(y_test,y_pred)
```

```python
model4=RandomForestClassifier()
model4.fit(x_train,y_train)
y_pred=model4.predict(x_test)
rfc=accuracy_score(y_test,y_pred)
classification_report(y_test,y_pred)
```

```python
model5=LogisticRegression()
model5.fit(x_train,y_train)
y_pred=model5.predict(x_test)
lg=accuracy_score(y_test,y_pred)
classification_report(y_test,y_pred)
```

```python
accu=['KNeighborsClassifier','SVC','DecisionTreeClassifier','RandomForestClassifier','LogisticRegression']
score=[knn,svc,dtc,rfc,lg]
plt.bar(accu,score)
```
